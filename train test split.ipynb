```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.DataFrame({'col1':range(100),'col2':range(100),'col3':range(100),'label':range(100)})
```

```python
def my_split(percentage,new_df,y):
    '''
    this is for train test split function practice
    '''
    
    lst = list(range(0,len(new_df)))
    random.seed('5')
    num_train = random.sample(lst,len(new_df))[:int((1-percentage)*len(new_df))]
    num_test = random.sample(lst,len(new_df))[:int(percentage*len(new_df))]
    
    X_train = new_df.iloc[num_train,:]
    X_test = new_df.iloc[num_test,:]
    
    y_train = y.iloc[num_train]
    y_test = y.iloc[num_test]
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = my_split(0.3,df.drop('label',axis = 1), y= df['label'])
```
